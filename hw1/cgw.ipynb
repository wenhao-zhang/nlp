### Parsing the sample files to generate grammar

We decided to throw out the given grammar files S1 and S2, because they do not match our intended tagging scheme.

We created our S1 Grammar rules based on parsing the devset.trees file. 

```python
from tree_to_grammar import grammar_from_file
grammar, terminals = grammar_from_file('devset.trees')
grammar
```

Because this parsed grammar has many rules that do not follow eCNF, we had to create a script to flatten the results into proper eCNF form.

For every rule that is too long, we generated a new intermediate rule composed of the last two atoms. This is repeated until all rules only have two symbols on the right hand side.

```python
from flatten_tree_results import flatten_grammar
flattened_grammar = flatten_grammar(grammar)
flattened_grammar
```

The previous step generated a lot of duplicate rules. This being the number of times some rule is used, we generated a new grammar file containing counts of the rules, which would help us generated our probabilties of each rule being used.

The result of this is our basic grammar rule set. 

```python
from grammar_rule_count import count_rules
counted_rules = count_rules(flattened_grammar)
counted_rules
```

### Fine Tuning

Using the grammar files generated with the above process with no adjustments did not produce good results.

We changed a few of our tagging assignments to make them more accurate based on our review and help bring down our cross entropy.

- Some interjections were assigned to being verbs or nouns, so we had to update some to be the proper tag UH. 
- Some rules generated sentences that were single brackets or other nonsensical single tokens. We chose to delete these.
- We also deleted rules that would not be appropriate for sentences from S2, such as commas being the end of sentences, sentences starting with starting with posessives (' and 's).

Our next step involved iteratively generating sentences and reviewing them for any unusual output.

- Taking the sentences, generating the parse tree, and printing it with print_tree.py so we can see which rules are causing the issue.
- Disabing or editing the rules that are cause improper generation, such as tags leading to other tags that are not compatible.
  - Example: NP (noun phrase) to PUNC (e.g. ., !, ?), or ADJP (adjective phrase) to RB (adverb)
- Updating ambiguous rules to be unique.
  
This got us from over -10 to about -9.5

#### Problems

As we proceeded with fine tuning the vocab for our test set, we realized that it was worsening performance on other test sentences that we had available. It was being fine tuned to be too specific to our test data.

We chose to revert to an earlier set with a bit less fine tuning and rule modification, and try other approaches to finding improvements.
